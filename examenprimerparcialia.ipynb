1. Escribe un programa que pida el peso (en kg) y la estatura (en metros) y calcule el IMC con la fórmula: Los datos de peso y altura deben ser ingresados por teclado

$$IMC=\frac{Peso}{estatura^2}$$
	​


```python
# Ingrese los datos
peso= float(input("Ingrese su peso en kilogramos (Kg):"))
estatura=float(input("Ingrese su estatura en metros(m):"))

# Cálculo del IMC
IMC=peso/(estatura*estatura)

# Resultado
print("Su índice de Masa Corporal (IMC) es:", IMC)
```

Escribe un programa que pida las coordenadas de dos puntos en el plano:
$$(x_1,y_1) y (x_2 , y_2)$$

Y calcule la distacia entre ellos
$$\sqrt{(x_2-x_1)^2+(y_2-y_1)^2}$$

Los datos x1,x2,y2,y2 son ingresados por teclado. Pedir al profesor ayuda para utilizar la función sqrt de la raiz


```python
import math
a=9
b=math.sqrt(a)


# Ingresar datos de coordenadas
x1= float(input("PUNTO 1 / Ingrese x1: "))
y1= float(input("PUNTO 1 / Ingrese y1: "))
x2= float(input("PUNTO 2 / Ingrese x2: "))
y2= float(input("PUNTO 2 / Ingrese y2: "))

# Calcular distancia
distancia= math.sqrt((x2-x1)*(x2-x1)+(y2-y1)*(y2-y1))

# Resultado

print("La distancia entre los dos puntos es:", distancia)
```
