# Comentarios

### ¿Qué son?

Texto contenido en ficheros de Python que es ignorado por el interprete; es decir no es ejecutado.

### ¿Cúal es su utilidad?

* Se puede utilizar para documentar código y hacerlo más legible.

* Preferiblemente trataremos de hacer fácil de entender y que necesite pocos comentarios, en lugar de vernos forzados a recurrir a los comentarios para explicar el código.

### Tipos de comentarios

**Comentarios de una línea**

Texto precedido por '#'
Se suele usar para documentar expresiones sencillas



```python
# Esto es una instruccion print
print("Hello word") # Esto es una instruccion print
```

**Comentarios de varias línea**

* Texto encapsulado en triple comillas (que pueden ser simples como dobles)
* Se suele usar para documentar bloques de código más significativo

```python
# creo una funcion de multiplicación
def producto(x, y):
    '''
    Esta función recibe dos números como parámetros y devuelve como resultado el producto de los mismos
     '''
    #creo una variable donde guardo el producto de los números recibidos
    c = x * y
    return c
```

```python
# Llamo a la función
producto(3, 7)
```

# Literales, variables y tipos de datos básicos

De forma muy genérica al ejecutarse un programa en Python se realizan *operaciones* sobre *objetos*

Estos dos términos son fundamentale.

*Objetos*: Cualquier tipo de datos (números, caracteres o datos más complejos).
* *Operaciones*: Cómo manipulamos estos datos. +,-,*,/

  Ejemplo:

```python
4 * 3
```

## Literales

* Python tienen una serie de tipos de datos integrados en el propio lenguaje.
* Los literales son expresiones que generan objetos de estos tipos.
* Estos objetos, según su tipo pueden ser
  * Simples o compuestos
  * Mutables o inmutables


### Literales simples

* Enteros
* Decimales o punto flotante
* Booleanos

```python
print(4)              #número entero
print(4.3)            #número en coma flotante
print('Hello word!')  #objeto de tipo string
print(False)          #Objeto de tipo booleano
```
