## Literales

* Python tienen una serie de tipos de datos integrados en el propio lenguaje.
* Los literales son expresiones que generan objetos de estos tipos.
* Estos objetos, según su tipo pueden ser
  * Simples o compuestos
  * Mutables o inmutables


### Literales simples

* Enteros
* Decimales o punto flotante
* Booleanos

```python
print(4)              #número entero
print(4.3)            #número en coma flotante
print('Hello word!')  #objeto de tipo string
print(False)          #Objeto de tipo booleano
```

### Literales compuestos

* Tuplas
* Listas
* Diccionarios
* Conjuntos

```python
print([1, 2, 3, 3]) # lista - mutable
print({'Nombre' : 'John Doe', "edad": 30}) # Diccionario - mutable
print({1, 2, 3, 3}) # Conjunto - mutable
print((4, 5)) # tupla - inmutable
2, 4 # tupla
```

## Variables

* Referencias a objetos.
* Las variables y los objetos se almacenan en diferentes zonas de memoria.
* Las variables siempre referencian a objetos y nunca a otras variables.
* Objetos sí que pueden referenciar a otros objetos. Ejemplo: listas.
* Sentencia de asignación:
<nombre_variable> '=' <objeto>

```python
# # Asignación de variables
a = 5
print(a)
```

```python
a = 1 # entero
b = 4.0 # coma flotante
c = "ITQ" # string
d = 10 + 1j # numero complejo
e = True #False # boolean
f = None # None
# visualizar valor de las variables y su tipo
print(a)
print(type(a))
print(b)
print(type(b))
print(c)
print(type(c))
print(d)
print(type(d))
print(e)
print(type(e))
print(f)
print(type(f))
```

momentos diferentes de la ejecución del programa, a objetos de diferente tipo.


```python
a = 3
print(a)
print(type(a))
a = 'Pablo García'
print(a)
print(type(a))
a = 4.5
print(a)
print(type(a))
```

##### Elaborar un programa que permita calcular el volumen de un cilindro, solicitando al usuario que ingrese los valores por teclado: radio y altura del cilindro. El resultado debe mostrarse en pantalla, No se debe utilizar condicionales, ni funciones matemáticas, solo operaciones aritméticas básicas y lectura de datos por teclado. Fórmula del volumen:

$$V = pi * r^ 2*h$$

r= radio h= altura que se ingrese por teclado

```python
#Solución

# Programa para calcular el volumen de un cilindro

pi = 3.1416

r = float(input("Ingrese el radio del cilindro: "))
h = float(input("Ingrese la altura del cilindro: "))

V = pi * r * r * h

print("El volumen del cilindro es:", V)

```

##### Elaborar un programa que permita convertir una distancia medida en metros a centimetros, milimetros y kilometros. El usuaario debe ingresar el valor en metros y el programa debe mostrar las tres equivalencias correspondientes. No se debe utilizar condicionales ni funciones matemáticas, únicamente operaciones aritméticas básicas.


```python
#Solución

# Programa para convertir metros a centimetros, milimetros y kilometros

m = float(input("Ingrese la distancia en metros: "))

cm = m * 100
mm = m * 1000
km = m / 1000

print("Distancia en centimetros:", cm, "cm")
print("Distancia en milimetros:", mm, "mm")
print("Distancia en kilometros:", km, "km")
```

## 3 Tipos de datos compuestos (colecciones)

### 3.1 Listas

* Una colección de objetos.
* Mutables.
* Tipos arbitrarios heterogeneos.
* Puede contener duplicados.
* Notienen tamaño fijo. Pueden contener tantos elementos como quepan en memoria.
* Los elementos van ordenados por posición.
* Se acceden usando la sintaxis: [index].
* Los índices van de 0 a n-1, donde n es el número de elementos de la lista.
* Son un tipo de Secuencia, al igual que los strings; por lo tanto, el orden (es decir, la posición
de los objetos de la lista) es importante.
* Soportan anidamiento.
* Son una implementación del tipo abstracto de datos: Array Dinámico.

#### Operaciones con listas

```python
letras = ['a', 'b', 'c', 'd']
palabras = 'Hola mundo como estas'.split()
numeros = list(range(7))
print(letras)
print(palabras)
print(numeros)
print(type(numeros))
```
