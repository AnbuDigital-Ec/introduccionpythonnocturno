## 1 Estructuras de control 

1.1 Selección (sentencias condicionales)
* Selección de una de varias alternativas en base a alguna condición.
* Indentación para estructurar el código
* Importante el símbolo ‘:’


```python
 # # Asigno el bvalo uno a la variable x
x = int(input("Ingrese el valor de x"))

# Implemento una condicional para saber si x es mayor a 0
if x > 0:
    print('Valor positivo')
else:
    print('Valor no positivo')
```

* Las condiciones son expresiones que se evalúan a True o False.
* Operadores de comparación, lógicos, de identidad y de pertenencia vistos anteriormente.

```python
x = 3
y = [1, 2, 3]

print(x > 0)

print(x > 0 and x < 10)
print(x is not y)
print(x in y)
```

* Se pueden introducir más ‘ramas’ en sentencias condicionales a través de la palabra clave elif.

```python
x = -3

if x > 0:
    
    print('Valor positivo')
elif x == 0:
    print('Valor nulo')
else:
    print('Valor negativo')
```

* Anidamiento.

```python
val = 120
if val > 0:
    if val < 100:
        print('Valor positivo')
    else:
        print('Valor muy positivo')
else:
    print('Valor negativo')
```

### EJERCICIOS sentencias condicionales

#### 1. Escribir un programa en Python que pregunte al usuario la edad por teclado y muestre en pantalla si es mayor de edad o no. Es mayor de edad cuando la edad es mayor o igual a 18


```python
# Ingrese la edad

edad = int(input("Ingrese su edad: "))

if edad >= 18:
    print("Es mayor de edad")
else:
    print("No es mayor de edad")
```

#### 2. Escribir un programa en Python donde el usuario ingrese 2 números por teclado y muestre por pantalla la división de los dos números. Si el divisor es cero, el programa debe mostrar el mensaje "No se puede dividir un valor para cero".

```python
# Ingreso de los dos números por teclado

num1= float(input("Ingrese el primer número: "))
num2= float(input("Ingrese el primer número: "))

# Verificación del divisor y realización de la división

if num2 != 0:
    print("El resultado de la división es:", num1 / num2)
else:
    print("No se puede dividir un valor para cero")
```

#### 3. Escribir un programa en Python que pida al usuario un numero por teclado y muestre en pantalla si el numero es par o impar.

```python
# Ingreso del número por teclado

numero = int(input("Ingrese un número: "))

# Verificación si el número es par o impar

if numero % 2 == 0:
    print("El número es par")
else:
    print("El número es impar")
```

## 2 Switch

```python
a = -6

match a:
    case _ if a > 0:
        print('positive')
    case _ if a == 0:
        print('zero')
    case _ if a < 0:
        print('negative')
```

```python
a = 'l' #week day

match a:
    case 'L':
        print("Lunes")
    case 'M':
        print("Martes")
    case 'X':
        print("Miercoles")
    case _:
        print("Wrong Day")
```

### Ejercicio

#### 1. Crear un programa en Python donde el usuario ingrese por teclado, el tipo de operaación aritmetica a realizar, ejemplo suma, resta, multiplicacion y division, ademas el usuario debe ingresar dos numeros enteros y realizarr la operacion que ingreso el usuario. Para el programa solo debe utilizar el condicional match

```python
# ingreso de datos
operacion = input("Ingrese la operación que desea realizar puede ser: suma, resta, multiplicacion, division ")
num1= int(input("Ingrese el primer número entero: "))
num2= int(input("Ingrese el segundo número entero: "))

# Uso del condicional match
match operacion:
    case "suma":
        print("Resultado:", num1 + num2)
    case "resta":
        print("Resultado:", num1 - num2)
    case "multiplicación":
        print("Resultado:", num1 * num2)
    case "division":
        if num2 != 0:
            print("Resultado", num1 / num2)
        else:
            print("Error: no se puede dividir para cero")
    case _:  
        print("Operación no válida")
        
```
