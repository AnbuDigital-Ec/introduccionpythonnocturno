Crear un programa en Python donde el usuario ingrese por teclado, el tipo de operaación aritmetica a realizar, ejemplo suma, resta, multiplicacion y division, ademas el usuario debe ingresar dos numeros enteros y realizarr la operacion que ingreso el usuario. Para el programa solo debe utilizar el condicional match

```python
# ingreso de datos
operacion = input("Ingrese la operación que desea realizar puede ser: suma, resta, multiplicacion, division ")
num1= int(input("Ingrese el primer número entero: "))
num2= int(input("Ingrese el segundo número entero: "))

# Uso del condicional match
match operacion:
    case "suma":
        print("Resultado:", num1 + num2)
    case "resta":
        print("Resultado:", num1 - num2)
    case "multiplicación":
        print("Resultado:", num1 * num2)
    case "division":
        if num2 != 0:
            print("Resultado", num1 / num2)
        else:
            print("Error: no se puede dividir para cero")
    case _:  
        print("Operación no válida")
        
```
