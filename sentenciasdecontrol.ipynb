##### Ejercicio 1: Desarrollar un programa en python donde se pida al usuario un número entero positivo y muestre por pantalla todos los números impares desde el 1 hasta ese número separado por comas

```python
# Programa que muestra los números impares desde 1 hasta un número dado

# Pido al usuario un número entero
n = int(input("Ingrese un número entero positivo: "))

# Verifico que el número sea positivo
if n <= 0:
    print("Error: debe ingresar un número entero positivo mayor que cero.")
else:
    impares = []  # Lista para almacenar los números impares

    # Recorro desde 1 hasta n
    for i in range(1, n + 1):
        if i % 2 != 0:          # Verifico si el número es impar
            impares.append(str(i))  # Convierto a texto y lo agrego a la lista

    # Muestro los números separados por comas
    print("Números impares desde 1 hasta", n, ":")
    print(", ".join(impares))
```

### Las sentencias break, continue y else
#### Break y continue
* Solo tienen sentido dentro de bucles.
* Break permite terminar el bucle por completo.
* Continue permite saltar a la siguiente iteración, continuando con el bucle.
* Pueden aparecer en cualquier parte de un bucle, pero normalmente aparecen dentro de sentencias condicionales (if).

```python
# # Ejemplo: break
rating_to_find = 4.2
movie_ratings = [4.3, 2.5, 1.7, 4.2, 3.8, 3.3, 4.5]

for rating in movie_ratings:
    print(rating)
    if rating == rating_to_find:
        print("Found")
        break
print(rating)
```

```python
for i in range(10):
    if i % 2 == 0:
        continue # Si el número es par, salta a la siguiente iteración␣
    print(f'Odd number {i}')
```

* Observa como la sentencia continue te puede ayudar a reducir el número de niveles de
anidamiento.
* Sin continue el anterior ejemplo sería:

```python
for i in range(10):
    if i % 2 != 0:
        print('Numero impar:', end=' ')
        print(i)
```

```python
for i in range(5):
    for a in range(2):
        print(f"{i} - {a}")
        if i == 3 and a == 0:
            break
```

#### Else
* Los bucles pueden tener una sensencia else.
* Resulta poco intuitiva para muchos programadores porque esta sintaxis no existe en otros
lenguajes.
* Se ejecuta cuando el bucle termina con normalidad; es decir, cuando no termina a causa de
un break.

```python
lista = [60,60,30,60]
element_to_find = 60

for element in lista:
    if element == element_to_find:
        print("found")
        break
else:
    print("not found")
```

```python
# with flag
lista = [10,30,50,40]
element_to_find = 30
found = False

for element in lista:
    if element == element_to_find:
        found = True
        break
if found:
    print("dato encontrado")
else:
    print("not found")
```

### List Comprehensions
* Permiten construir listas a través de la ejecución repetida (sentencia for) de una expressión
para cada item de un objeto iterable.
* Van entre ‘[’ y ’]’. Esto es indicativo de que estamos construyendo una lista.
  
Sintaxis:<br>
[<expression> for <item> in <iterable>]
Ejemplo: repetir los carácteres de un string.

```python
[char*3 for char in "Enrique"]
```

```python
lista = []
for char in 'Enrique':
    lista.append(char)
print(lista)
```

Comúnmente, el item de la sentencia for aparecerá en la expresión principal, pero eso no es obligatorio.

```python
[2 for _ in 'Enrique']
```

#### 4.0.1 Versión extendida

Se puede especificar un filtro (sentencia if ) para obtener únicamente los elementos que cumplan
cierta condición.

Sintaxis:<br>
[<expression> for <item> in <iterable> if <condition>]
Ejemplo: obtener números pares:

```python
[y for y in range(9) if y % 2 == 0]
```

* Las list comprehension no son realmente requeridas, ya que siempre podemos escribir un bucle
equivalente.

```python
pares = []
for x in range(9):
    if x % 2 == 0:
        pares.append(x)
print(pares)
```

#### 4.0.2 Versión completa
La sentencia if de una list comprehension también puede contener una expresión alternativa.

Sintaxis:<br>
[<expression_1> if <condition> else <expression_2> for <item> in <iterable>]
Ejemplo: poner a cero los números pares.

```python
[0 if x % 2 == 0 else x for x in range(9)]
```

```python
[x**2 if x > 2 else x for x in range(-4,5) if x > 0]
```

```python
lista = []
for x in range(-4,5):
    if x > 0:
        if x > 2:
            lista.append(x**2)
        else:
            lista.append(x)
print(lista)
```

```python
[x**2 if x > 0 else 100 if x==0 else x for x in range(-4,5)]
```

```python
lista = []
for x in range(-4,5):
    if x > 0:
        lista.append(x**2)
    elif x == 0:
        lista.append(100)
    else:
        lista.append(x)
print(lista)
```

#### 4.0.3 Anidamiento
Las list comprehensions soportan anidamiento en sus expresiones.

dadas.

```python
lista_1 = [1, 2, 3]
lista_2 = [4, 5]
[(x, y) for x in lista_1 for y in lista_2]
```

#### 4.0.4 Pros y contras
* Principales ventajas de las comprehensions en comparación a un bucle convencional:

– Expresión compacta y legible, si estás familiarizado con la sintaxis.<br>
– Mejor rendimiento.
* Desventaja: no escalan bien. Una list comprehension se puede convertir rápidamente en una
expresión difícil de entender.

1. Escribe un programa que calcule la suma de todos los elementos de una lista dada. La lista
sólo puede contener elementos numéricos.

```python
# Función que calcula la suma de los elementos de una lista
def suma_lista(lista):
    suma = 0  # Acumulador

    for elemento in lista:
        suma += elemento

    return suma


# Lista de ejemplo (solo números)
numeros = [5, 10, 3, 7, 2]

# Llamada a la función
resultado = suma_lista(numeros)

# Mostrar el resultado
print("La suma de los elementos de la lista es:", resultado)
```

2. Dada una lista con elementos duplicados, escribir un programa que muestre una nueva lista
con el mismo contenido que la primera pero sin elementos duplicados. Para este ejercicio, no
puedes hacer uso de objetos de tipo ‘Set’.

```python
# Función que elimina los elementos duplicados de una lista
def eliminar_duplicados(lista):
    lista_sin_duplicados = []

    for elemento in lista:
        if elemento not in lista_sin_duplicados:
            lista_sin_duplicados.append(elemento)

    return lista_sin_duplicados


# Lista de ejemplo con elementos duplicados
lista_original = [1, 2, 3, 2, 4, 1, 5, 3]

# Llamada a la función
resultado = eliminar_duplicados(lista_original)

# Mostrar resultados
print("Lista original:", lista_original)
print("Lista sin duplicados:", resultado)
```

3. Escribe un programa que construya un diccionario que contenga un número (entre 1 y n) de
elementos de esta forma: (x, x*x). Ejemplo: para n = 5, el diccionario resultante sería {1: 1,
2: 4, 3: 9, 4: 16, 5: 25}

```python
# Función que construye un diccionario con pares (x, x*x)
def construir_diccionario(n):
    diccionario = {}

    for x in range(1, n + 1):
        diccionario[x] = x * x

    return diccionario


# Entrada de datos
n = int(input("Ingrese un número entero positivo: "))

# Validación
if n <= 0:
    print("Error: el número debe ser mayor que cero.")
else:
    # Llamada a la función
    resultado = construir_diccionario(n)

    # Mostrar resultado
    print("Diccionario resultante:")
    print(resultado)
```

que indique el éxito o el fracaso de la búsqueda. En caso de éxito, también se mostrará por
pantalla la palabra encontrada.

```python
# Función que busca una palabra que empiece por 'a' y tenga más de 9 caracteres
def buscar_palabra(lista_palabras):
    for palabra in lista_palabras:
        if palabra.lower().startswith('a') and len(palabra) > 9:
            return palabra  # Termina en el momento exacto de encontrarla
    return None


# Lista de palabras de ejemplo
palabras = [
    "casa",
    "avion",
    "automovil",
    "astronauta",
    "biblioteca",
    "arbol"
]

# Llamada a la función
resultado = buscar_palabra(palabras)

# Mostrar resultado
if resultado:
    print("Búsqueda exitosa.")
    print("Palabra encontrada:", resultado)
else:
    print("Búsqueda fallida. No se encontró ninguna palabra que cumpla las condiciones.")
```

5. Dada una lista L de números positivos, escribir un programa que muestre otra lista (ordenada)
que contenga todo índice i que cumpla la siguiente condición: L[i] es múltiplo de 3. Por ejemplo, dada la lista L = [3,5,13,12,1,9] el programa mostrará la lista [0,3,5] dado que L[0],
L[3] y L[5] son, respectivamente, 3, 12 y 9, que son los únicos múltiplos de 3 que hay en L.

```python
# Lista de números positivos
L = [3, 5, 13, 12, 1, 9]

# Lista para almacenar los índices
indices = []

# Recorrer la lista usando índices
for i in range(len(L)):
    if L[i] % 3 == 0:
        indices.append(i)

# Mostrar el resultado
print("Índices cuyos valores son múltiplos de 3:", indices)
```

6. Dado un diccionario cuyos elementos son pares de tipo string y numérico (es decir, las claves
son de tipo ‘str’ y los valores son de tipo ‘int’ o ‘float’), escribe un programa que muestre
por pantalla la clave cuyo valor asociado representa el valor númerico más alto de todo el
diccionario. Por ejemplo, para el diccionario {‘a’: 4.3, ‘b’: 1, ‘c’: 7.8, ‘d’: -5} la respuesta
sería ‘c’, dado que 7.8 es el valor más alto de los números 4.3, 1, 7.8 y -5.

```python
# Diccionario dado
diccionario = {'a': 4.3, 'b': 1, 'c': 7.8, 'd': -5}

# Inicializamos variables
mayor_valor = None
clave_mayor = None

# Recorremos el diccionario
for clave, valor in diccionario.items():
    if mayor_valor is None or valor > mayor_valor:
        mayor_valor = valor
        clave_mayor = clave

# Mostramos el resultado
print("La clave con el valor numérico más alto es:", clave_mayor)
```

7. Dada la lista a = [2, 4, 6, 8] y la lista b = [7, 11, 15, 22], escribe un programa que itere las
listas a y b y multiplique cada elemento de a que sea mayor que 5 por cada elemento de b que
sea menor que 14. El programa debe mostrar los resultados por pantalla.

```python
# Listas dadas
a = [2, 4, 6, 8]
b = [7, 11, 15, 22]

# Recorremos la lista a
for elemento_a in a:
    if elemento_a > 5:
        
        # Recorremos la lista b
        for elemento_b in b:
            if elemento_b < 14:
                resultado = elemento_a * elemento_b
                print(f"{elemento_a} x {elemento_b} = {resultado}")
```

10. Escribe una list comprehension que construya una lista con los números enteros positivos de
una lista de números dada. La lista original puede incluir números de tipo float, los cuales
deben ser descartados.

```python
# Lista original
lista = [3, -2, 4.5, 7, 0, -1, 6.0, 9, 2.3]

# List comprehension para enteros positivos
lista_enteros_positivos = [x for x in lista if isinstance(x, int) and x > 0]

# Mostrar resultado
print("Lista original:", lista)
print("Lista de enteros positivos:", lista_enteros_positivos)
```

13. Escribe una dictionary comprehension que construya un diccionario que incluya los tamaños
de cada palabra en una frase dada. Ejemplo: el resultado para la frase “Soy un ser humano”
será {‘Soy’: 3, ‘un’: 2, ‘ser’: 3, ‘humano’: 6}

```python
# Solicitar la frase al usuario
frase = input("Ingrese una frase: ")

# Dictionary comprehension para obtener el tamaño de cada palabra
diccionario_tamanios = {palabra: len(palabra) for palabra in frase.split()}

# Mostrar resultado
print(diccionario_tamanios)
```

14. Escribe una list comprehension que construya una lista que incluya todos los números del 1
al 10 en orden. La primera mitad se mostrarán en formato numérico; la segunda mitad en
texto. Es decir, el resultado será: [1, 2, 3, 4, 5, ‘seis’, ‘siete’, ‘ocho’, ‘nueve’, ‘diez’].

```python
# Diccionario de números en texto
numeros_texto = {
    6: "seis",
    7: "siete",
    8: "ocho",
    9: "nueve",
    10: "diez"
}

# List comprehension solicitada
lista_resultado = [x if x <= 5 else numeros_texto[x] for x in range(1, 11)]

# Mostrar resultado
print(lista_resultado)
```
